```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
URL = "https://covid19.mhlw.go.jp/public/opendata/newly_confirmed_cases_daily.csv"
df = pd.read_csv(URL,parse_dates=['Date'])
df.tail(1)
#df = pd.read_csv('./newly_confirmed_cases_daily.csv', parse_dates=['Date'])
fig, ax = plt.subplots(figsize=(20, 15))
formatter = mdates.DateFormatter('%y/%m')
ax.xaxis.set_major_formatter(formatter)
ax.plot(df['Date'], df['Tokyo'],label="Tokyo")
ax.plot(df['Date'], df['Osaka'],label="Osaka")
ax.plot(df['Date'], df['Okinawa'],label="Okinawa")
#ax.plot(df['Date'], df['Hiroshima'],label="Hiroshima")
#ax.plot(df['Tokyo'], df['Osaka'],label="Hiroshima")
ax.grid(True)
plt.legend(['Tokyo','Osaka','Okinawa'])
#plt.legend(['Osaka','Okinawa','Hiroshima'])
#plt.legend(['Okinawa','Hiroshima'])
plt.yscale('log')
plt.xticks(rotation=20)
plt.show()
```
