```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
URL = "https://covid19.mhlw.go.jp/public/opendata/newly_confirmed_cases_daily.csv"
df = pd.read_csv(URL,parse_dates=['Date'])
df.tail(1)
#df = pd.read_csv('./newly_confirmed_cases_daily.csv', parse_dates=['Date'])
fig, ax = plt.subplots(figsize=(20, 15))
formatter = mdates.DateFormatter('%y/%m')
ax.xaxis.set_major_formatter(formatter)
ax.plot(df['Date'], df['Tokyo'],'*',label="Tokyo")
ax.plot(df['Date'], df['Osaka'],'+',label="Osaka")
ax.plot(df['Date'], df['Okinawa'],'o',label="Okinawa")
ax.grid(True)
plt.legend(['Tokyo','Osaka','Okinawa'])
plt.yscale('log')
plt.ylim(1e-1,1e4)
plt.xticks(rotation=20)
plt.show()
```
